```python
df= df[df['Area']<= 5000]
```

```python
df=df[df['BHK']<=5]
```

```python
df['Per_Sqft'].fillna(method='ffill',inplace=True)
```

```python
df['Per_Sqft'].fillna(6667.0,inplace=True)
```

```python
df['Parking'].value_counts()
```

```python
df['Parking'].fillna(df['Parking'].mode(),inplace=True)
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
df=df.drop(columns=['Locality'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
furnishing_encoder=LabelEncoder()
status_encoder=LabelEncoder()
transaction_encoder=LabelEncoder()
type_encoder=LabelEncoder()
```

```python
df['Furnishing']= furnishing_encoder.fit_transform(df['Furnishing'])
df['Status']= status_encoder.fit_transform(df['Status'])
df['Transaction']= transaction_encoder.fit_transform(df['Transaction'])
df['Type']= type_encoder.fit_transform(df['Type'])



```

```python
df.corr()['Price']
```

```python
df.drop(columns=['Furnishing','Parking'],inplace=True)
```

```python
df.head()
```

```python
df['Bathroom'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first')
```

```python
X_trans=ohe.fit_transform(df[['BHK','Bathroom']]).toarray()
```

```python
X=df.drop(columns=['BHK','Bathroom','Price']).values
```

```python
X=np.hstack((X,X_trans))
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X=scaler.fit_transform(X)
```

```python
y=df['Price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
